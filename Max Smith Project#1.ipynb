```python
#To start, we to need to insert out packages. 
import pandas as pd


#We then need to load our data and make sure its the correct resource.
air_data = pd.read_csv('AirQuality_Daily_StudentVersion.csv')
```

```python
#We then need to turn our data into a data frame, this is the code how

air_data = pd.DataFrame(air_data)
```

```python
#Next we make sure the VOC data is grouped by sensor name
VOC_grouped = air_data.groupby('sensor.name')

#We then need to tell the code what statistics to give me a summary of voc
air_data_VOC = VOC_grouped.agg(max = ('voc','max'),
                               min = ('voc','min'),
                               mean = ('voc','mean'),
                               median = ('voc','median'))
#Now that we have those satistics we will then seperate them out by average and median

top5_avg_VOC = ( air_data_VOC.sort_values(by = 'mean', ascending = False))
top5_avg_VOC.head(5)

top5_med_VOC = ( air_data_VOC.sort_values(by = 'median', ascending = False))

top5_med_VOC.head(5)

```

```python
#Next we will group the pm2.5_atm data by sensor name

pm25_grouped = air_data.groupby('sensor.name')

#We will now ask python to give me a summary of pm2.5

air_data_pm25 = pm25_grouped.agg(max = ('pm2.5_atm','max'),
                                   min = ('pm2.5_atm','min'),
                                   mean = ('pm2.5_atm','mean'),
                                   median = ('pm2.5_atm','median'))


#Now that we have those satistics we will then seperate them out by average and median

top5_avg_pm25 = (air_data_pm25.sort_values(by = 'mean', ascending = False))

top5_avg_pm25.head(5)

top5_med_pm25 = ( air_data_pm25.sort_values(by = 'median', ascending = False))

top5_med_pm25.head(5)

```

```python
# This next line of code will group the pm10.0_atm data by sensor name.
pm10_grouped = air_data.groupby('sensor.name')

air_data_pm10 = pm10_grouped.agg(max = ('pm10.0_atm','max'),
                                 min = ('pm10.0_atm','min'),
                                 mean = ('pm10.0_atm','mean'),
                                 median = ('pm10.0_atm','median'))

top5_avg_pm10 = (air_data_pm10.sort_values(by = 'mean', ascending = False))

top5_avg_pm10.head(5)

top5_med_pm10 = (air_data_pm10.sort_values(by = 'median', ascending = False))

top5_med_pm10.head(5)

```

```python
#This next line will give the temperature assigned temperature levels
def assign_temperature(temperature):
    if temperature < 32:
        return 'Below Freezing'
    elif 32 <= temperature <= 50:
        return 'Cool'
    elif 50 <= temperature <= 70:
        return 'Warm'
    elif 70 <= temperature <= 150:
        return 'Hot'

#We then need a line to apply the function

air_data['temperature_level'] = air_data['temperature'].apply(assign_temperature)

#Next we will group the data by temperature level and sensor name

temperature_sensor = (air_data.groupby(['sensor.name','temperature_level']).size())

temperature_sensor
```

```python
#This next line will give the data assigned humidity levels
def assign_humidity(humidity):
    if humidity < 50:
        return 'Low'
    elif 50 <= humidity <= 80:
        return 'High'
    elif 80 <= humidity <= 100:
        return 'Very High'
    

#We then need a line to apply the function

air_data['humidity_level'] = air_data['humidity'].apply(assign_humidity)

#Next we will group the data by humidity level and sensor name

humidity_sensor = (air_data.groupby(['sensor.name','humidity_level']).size())

humidity_sensor
```

```python
# We will now group pm2.5_atm and give them assigned levels
def assign_pm25(pm25):
    if pm25 < 10:
        return 'good'
    elif 10 <= pm25 <= 40:
        return 'Moderate'
    elif 40 <= pm25 <= 60:
        return 'Unhealthy for Sensitive Groups'
    elif 60 <= pm25 <= 150:
        return 'Unhealthy'
    elif 150 <= pm25 <= 250:
        return 'Very Unhealthy'
    elif 250 <= pm25 <= 1000:
        return 'Hazardous'

#We then need to apply this aplication
air_data['pm2.5_AQI'] = air_data['pm2.5_atm'].apply(assign_pm25)
air_data

#Next we will group the data by pm2.5_atm levels and sensor name

pm25_sensor = (air_data.groupby(['sensor.name','pm2.5_AQI']).size())

pm25_sensor
               
```

```python
# We will now group pm10.0_atm and give them assigned levels
def assign_pm10(pm10):
    if pm10 < 55:
        return 'good'
    elif 55 <= pm10 <= 155:
        return 'Moderate'
    elif 155 <= pm10 <= 255:
        return 'Unhealthy for Sensitive Groups'
    elif 255 <= pm10 <= 355:
        return 'Unhealthy'
    elif 355 <= pm10 <= 425:
        return 'Very Unhealthy'
    elif 425 <= pm10 <= 1000:
        return 'Hazardous'

#We then need to apply this aplication
air_data['pm10.0_AQI'] = air_data['pm10.0_atm'].apply(assign_pm10)
air_data

#Next we will group the data by pm2.5_atm levels and sensor name

pm10_sensor = (air_data.groupby(['sensor.name','pm10.0_AQI']).size())

pm10_sensor
```
